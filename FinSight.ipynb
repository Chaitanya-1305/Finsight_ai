```python
# ============================================================================
# STEP 1B: Extract Text from All PDFs in Your Folders
# ============================================================================

import os
import json
import pdfplumber
from pathlib import Path

# YOUR FOLDER PATHS (matching your Drive structure)
annual_path = "/content/drive/MyDrive/Annual Reports"
quarterly_path = "/content/drive/MyDrive/Quarterly Reports"
output_base = "/content/drive/MyDrive/processed_data"

# Create output directory
os.makedirs(output_base, exist_ok=True)
os.makedirs(f"{output_base}/texts", exist_ok=True)
os.makedirs(f"{output_base}/metadata", exist_ok=True)

print("=" * 70)
print("STEP 1B: EXTRACTING TEXT FROM PDFs")
print("=" * 70)

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    """Extract all text from a PDF file."""
    text = ""
    try:
        with pdfplumber.open(pdf_path) as pdf:
            for page in pdf.pages:
                page_text = page.extract_text()
                if page_text:
                    text += page_text + "\n"
    except Exception as e:
        print(f"  ERROR: {str(e)[:50]}")
    return text

# Process Annual Reports
print("\nðŸ“„ Processing Annual Reports...")
annual_files = [f for f in os.listdir(annual_path) if f.endswith('.pdf')]
annual_metadata = []

for idx, pdf_file in enumerate(annual_files, 1):
    pdf_path = os.path.join(annual_path, pdf_file)
    company_name = pdf_file.replace('.pdf', '')

    print(f"  [{idx}/{len(annual_files)}] {company_name}...", end="")

    # Extract text
    text = extract_text_from_pdf(pdf_path)

    # Save text file
    text_file = f"{output_base}/texts/{company_name}_annual.txt"
    with open(text_file, 'w', encoding='utf-8') as f:
        f.write(text)

    # Record metadata
    annual_metadata.append({
        'company': company_name,
        'type': 'annual',
        'file': pdf_file,
        'text_length': len(text),
        'status': 'success' if len(text) > 0 else 'empty'
    })

    print(f" âœ“ ({len(text)} chars)")

# Process Quarterly Reports
print("\nðŸ“Š Processing Quarterly Reports...")
quarterly_files = [f for f in os.listdir(quarterly_path) if f.endswith('.pdf')]
quarterly_metadata = []

for idx, pdf_file in enumerate(quarterly_files, 1):
    pdf_path = os.path.join(quarterly_path, pdf_file)
    company_name = pdf_file.replace('.pdf', '')

    print(f"  [{idx}/{len(quarterly_files)}] {company_name}...", end="")

    # Extract text
    text = extract_text_from_pdf(pdf_path)

    # Save text file
    text_file = f"{output_base}/texts/{company_name}_quarterly.txt"
    with open(text_file, 'w', encoding='utf-8') as f:
        f.write(text)

    # Record metadata
    quarterly_metadata.append({
        'company': company_name,
        'type': 'quarterly',
        'file': pdf_file,
        'text_length': len(text),
        'status': 'success' if len(text) > 0 else 'empty'
    })

    print(f" âœ“ ({len(text)} chars)")

# Save metadata
metadata = annual_metadata + quarterly_metadata
metadata_df = pd.DataFrame(metadata)
metadata_df.to_csv(f"{output_base}/metadata/extraction_metadata.csv", index=False)

print("\n" + "=" * 70)
print("âœ“ EXTRACTION COMPLETE!")
print("=" * 70)
print(f"Total PDFs processed: {len(annual_files) + len(quarterly_files)}")
print(f"Texts saved to: {output_base}/texts/")
print(f"Metadata saved to: {output_base}/metadata/extraction_metadata.csv")
print("\nSummary:")
print(metadata_df.groupby('type')['status'].value_counts())
```

```python
# ============================================================================
# STEP 1C: Segment Extracted Text into Financial Sections
# ============================================================================

import os
import json
import re

# Paths from previous step
output_base = "/content/drive/MyDrive/processed_data"
texts_path = f"{output_base}/texts"
segments_path = f"{output_base}/segments"

# Create output directory for segmented data
os.makedirs(segments_path, exist_ok=True)

print("=" * 70)
print("STEP 1C: SEGMENTING TEXT INTO FINANCIAL SECTIONS")
print("=" * 70)

def segment_text(text):
    """
    Segment financial report text into key sections.

    Returns a dictionary with sections and their content.
    """
    segments = {
        'ceo_letter': '',
        'mda': '',
        'risk_factors': '',
        'financial_statements': '',
        'full_text': text
    }

    text_lower = text.lower()

    # ===== CEO LETTER / CHAIRMAN'S MESSAGE =====
    ceo_keywords = [
        'letter to shareholder',
        'message from ceo',
        'chairman\'s message',
        'dear shareholder',
        'from the desk of',
        'message to stakeholder'
    ]

    for keyword in ceo_keywords:
        idx = text_lower.find(keyword)
        if idx != -1:
            # Extract ~3000 chars after keyword (CEO letter usually short)
            segments['ceo_letter'] = text[idx:idx + 3000]
            break

    # ===== MD&A (Management Discussion & Analysis) =====
    mda_keywords = [
        'management discussion and analysis',
        'md&a',
        'management\'s discussion',
        'operating performance',
        'financial performance'
    ]

    for keyword in mda_keywords:
        idx = text_lower.find(keyword)
        if idx != -1:
            # Extract ~5000 chars (MD&A is usually longer)
            segments['mda'] = text[idx:idx + 5000]
            break

    # ===== RISK FACTORS / RISKS & UNCERTAINTIES =====
    risk_keywords = [
        'risk factor',
        'risks and uncertainties',
        'risk management',
        'material risk',
        'going concern'
    ]

    for keyword in risk_keywords:
        idx = text_lower.find(keyword)
        if idx != -1:
            # Extract ~4000 chars
            segments['risk_factors'] = text[idx:idx + 4000]
            break

    # ===== FINANCIAL STATEMENTS =====
    financial_keywords = [
        'balance sheet',
        'statement of profit and loss',
        'consolidated financial statement',
        'standalone financial statement',
        'statement of cash flow'
    ]

    for keyword in financial_keywords:
        idx = text_lower.find(keyword)
        if idx != -1:
            # Extract ~3000 chars (just the header area)
            segments['financial_statements'] = text[idx:idx + 3000]
            break

    return segments

# Process all extracted text files
text_files = [f for f in os.listdir(texts_path) if f.endswith('.txt')]

print(f"\nSegmenting {len(text_files)} text files...\n")

for idx, text_file in enumerate(text_files, 1):
    text_path = os.path.join(texts_path, text_file)

    # Read extracted text
    with open(text_path, 'r', encoding='utf-8') as f:
        text = f.read()

    # Segment the text
    segments = segment_text(text)

    # Save segmented data as JSON
    segment_file = text_file.replace('.txt', '_segments.json')
    segment_path = os.path.join(segments_path, segment_file)

    with open(segment_path, 'w', encoding='utf-8') as f:
        json.dump(segments, f, indent=2, ensure_ascii=False)

    # Print progress
    company_name = text_file.replace('_annual.txt', '').replace('_quarterly.txt', '')
    ceo_len = len(segments['ceo_letter'])
    mda_len = len(segments['mda'])
    risk_len = len(segments['risk_factors'])
    fin_len = len(segments['financial_statements'])

    print(f"[{idx}/{len(text_files)}] {company_name}")
    print(f"         CEO: {ceo_len}  |  MD&A: {mda_len}  |  Risk: {risk_len}  |  Fin: {fin_len}")

print("\n" + "=" * 70)
print("âœ“ SEGMENTATION COMPLETE!")
print("=" * 70)
print(f"Total files segmented: {len(text_files)}")
print(f"Segmented data saved to: {segments_path}/")
print("\nSegment types created:")
print("  - CEO Letter / Chairman's Message")
print("  - MD&A (Management Discussion & Analysis)")
print("  - Risk Factors & Uncertainties")
print("  - Financial Statements Header")
print("\nNext: These segments will be labeled for training AI models!")
```

```python
# ============================================================================
# STEP 3: Build Baseline Rules for KPI Extraction
# ============================================================================

import re
import pandas as pd
import json
from pathlib import Path

# Paths
output_base = "/content/drive/MyDrive/processed_data"
segments_path = f"{output_base}/segments"
results_path = f"{output_base}/baseline_results"

# Create output directory
os.makedirs(results_path, exist_ok=True)

print("=" * 70)
print("STEP 3: BUILDING BASELINE KPI EXTRACTION RULES")
print("=" * 70)

# ============================================================================
# PART 3A: Define Regex Patterns for KPI Extraction
# ============================================================================

class KPIExtractor:
    """Extract KPIs from financial text using regex patterns."""

    def __init__(self):
        self.kpi_patterns = {
            'revenue': [
                r'revenue.*?([\$â‚¹]\s*[\d,\.]+\s*(?:billion|million|crore|lakh))',
                r'total income.*?([\$â‚¹]\s*[\d,\.]+\s*(?:billion|million|crore|lakh))',
                r'sales.*?([\$â‚¹]\s*[\d,\.]+\s*(?:billion|million|crore|lakh))',
            ],
            'net_profit': [
                r'net profit.*?([\$â‚¹]\s*[\d,\.]+\s*(?:billion|million|crore|lakh))',
                r'net income.*?([\$â‚¹]\s*[\d,\.]+\s*(?:billion|million|crore|lakh))',
                r'profit for the (?:year|period).*?([\$â‚¹]\s*[\d,\.]+\s*(?:billion|million|crore|lakh))',
            ],
            'eps': [
                r'earnings per share.*?([\d\.]+)',
                r'eps.*?([\d\.]+)',
            ],
            'roe': [
                r'return on equity.*?([\d\.]+)\s*%',
                r'roe.*?([\d\.]+)\s*%',
            ],
            'profit_margin': [
                r'profit margin.*?([\d\.]+)\s*%',
                r'net margin.*?([\d\.]+)\s*%',
            ],
            'current_ratio': [
                r'current ratio.*?([\d\.]+)',
                r'liquidity ratio.*?([\d\.]+)',
            ],
            'debt_to_equity': [
                r'debt.*?equity ratio.*?([\d\.]+)',
                r'debt.*?equity.*?([\d\.]+)',
            ],
        }

    def extract_kpis(self, text):
        """Extract KPIs from text using regex patterns."""
        extracted_kpis = {}
        text_lower = text.lower()

        for kpi_name, patterns in self.kpi_patterns.items():
            for pattern in patterns:
                matches = re.findall(pattern, text_lower, re.IGNORECASE)
                if matches:
                    extracted_kpis[kpi_name] = matches[0]  # Take first match
                    break  # Move to next KPI

        return extracted_kpis

# ============================================================================
# PART 3B: Test Extraction on All Segmented Files
# ============================================================================

print("\nðŸ” Testing KPI Extraction on all documents...\n")

extractor = KPIExtractor()
extraction_results = []

segment_files = [f for f in os.listdir(segments_path) if f.endswith('_segments.json')]

for idx, seg_file in enumerate(segment_files, 1):
    seg_path = os.path.join(segments_path, seg_file)

    # Load segmented data
    with open(seg_path, 'r', encoding='utf-8') as f:
        segments = json.load(f)

    # Extract KPIs from financial statements section
    fin_text = segments.get('financial_statements', '')

    if fin_text:
        kpis = extractor.extract_kpis(fin_text)
    else:
        kpis = {}

    # Record results
    company_name = seg_file.replace('_annual_segments.json', '').replace('_quarterly_segments.json', '')
    doc_type = 'annual' if 'annual' in seg_file else 'quarterly'

    result = {
        'company': company_name,
        'document_type': doc_type,
        'kpis_found': len(kpis),
        **kpis
    }
    extraction_results.append(result)

    # Print progress
    print(f"[{idx}/{len(segment_files)}] {company_name} ({doc_type})")
    print(f"         KPIs found: {len(kpis)}")
    if kpis:
        for kpi, value in list(kpis.items())[:3]:  # Show first 3
            print(f"           - {kpi}: {value}")

# ============================================================================
# PART 3C: Save Baseline Results
# ============================================================================

print("\n" + "=" * 70)
print("âœ“ BASELINE EXTRACTION COMPLETE!")
print("=" * 70)

# Convert to DataFrame and save
results_df = pd.DataFrame(extraction_results)
results_csv = f"{results_path}/kpi_extraction_baseline.csv"
results_df.to_csv(results_csv, index=False)

print(f"\nExtraction results saved to: {results_csv}")

# Statistics
print("\nðŸ“Š BASELINE EXTRACTION STATISTICS:")
print("-" * 70)
print(f"Total documents processed: {len(extraction_results)}")
print(f"Documents with KPIs found: {len([r for r in extraction_results if r['kpis_found'] > 0])}")
print(f"Average KPIs per document: {results_df['kpis_found'].mean():.2f}")

# Show which KPIs were most commonly found
kpi_columns = [col for col in results_df.columns if col not in ['company', 'document_type', 'kpis_found']]
print(f"\nKPIs found across documents:")
for kpi in kpi_columns:
    count = results_df[kpi].notna().sum()
    if count > 0:
        print(f"  - {kpi}: {count} documents ({count/len(results_df)*100:.1f}%)")

print("\n" + "=" * 70)
print("ðŸ“ WHAT THIS BASELINE SHOWS:")
print("=" * 70)
print("""
This baseline extraction:
âœ“ Uses simple regex rules to find KPIs
âœ“ Shows which KPIs are easy to extract
âœ“ Gives us a comparison point for ML models
âœ“ Identifies KPIs that need better patterns

Next: Step 4 will train actual ML models for:
  - Sentiment Analysis (FinBERT)
  - Risk Classification (Transformer model)
  - Advanced KPI Extraction (NER model)
""")
```

```python
# ============================================================================
# STEP 6: CREATE INFERENCE PIPELINE
# ============================================================================

import os
import json
import re
import torch
import pandas as pd
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import warnings
warnings.filterwarnings('ignore')

# Paths
output_base = "/content/drive/MyDrive/processed_data"
models_path = f"{output_base}/trained_models"
inference_path = f"{output_base}/inference_results"

os.makedirs(inference_path, exist_ok=True)

print("=" * 70)
print("STEP 6: CREATE INFERENCE PIPELINE")
print("=" * 70)

# ============================================================================
# PART A: Load Trained Models
# ============================================================================

print("\nðŸ¤– Loading trained models...\n")

# Load sentiment model
sentiment_model_path = f"{models_path}/finbert_sentiment_final"
sentiment_tokenizer = AutoTokenizer.from_pretrained(sentiment_model_path)
sentiment_model = AutoModelForSequenceClassification.from_pretrained(sentiment_model_path)

with open(f"{sentiment_model_path}/label_mapping.json", 'r') as f:
    sentiment_labels = json.load(f)

print("âœ“ Sentiment model loaded")

# Load risk model
risk_model_path = f"{models_path}/finbert_risk_final"
risk_tokenizer = AutoTokenizer.from_pretrained(risk_model_path)
risk_model = AutoModelForSequenceClassification.from_pretrained(risk_model_path)

with open(f"{risk_model_path}/label_mapping.json", 'r') as f:
    risk_labels = json.load(f)

print("âœ“ Risk model loaded")

# ============================================================================
# PART B: Define KPI Extraction Patterns
# ============================================================================

print("\nðŸ“Š Setting up KPI extraction patterns...\n")

kpi_patterns = {
    'revenue': [
        r'revenue.*?([\$â‚¹]\s*[\d,\.]+\s*(?:billion|million|crore|lakh))',
        r'total income.*?([\$â‚¹]\s*[\d,\.]+\s*(?:billion|million|crore|lakh))',
    ],
    'net_profit': [
        r'net profit.*?([\$â‚¹]\s*[\d,\.]+\s*(?:billion|million|crore|lakh))',
        r'net income.*?([\$â‚¹]\s*[\d,\.]+\s*(?:billion|million|crore|lakh))',
    ],
    'eps': [
        r'earnings per share.*?([\d\.]+)',
        r'eps.*?([\d\.]+)',
    ],
    'roe': [
        r'return on equity.*?([\d\.]+)\s*%',
        r'roe.*?([\d\.]+)\s*%',
    ],
    'profit_margin': [
        r'profit margin.*?([\d\.]+)\s*%',
        r'net margin.*?([\d\.]+)\s*%',
    ],
}

print("âœ“ KPI patterns ready")

# ============================================================================
# PART C: Define Inference Functions
# ============================================================================

class FinancialAnalysisPipeline:
    """Complete pipeline for financial document analysis."""

    def __init__(self, sentiment_tokenizer, sentiment_model, risk_tokenizer, risk_model,
                 sentiment_labels, risk_labels, kpi_patterns):
        self.sentiment_tokenizer = sentiment_tokenizer
        self.sentiment_model = sentiment_model
        self.risk_tokenizer = risk_tokenizer
        self.risk_model = risk_model
        self.sentiment_labels = sentiment_labels
        self.risk_labels = risk_labels
        self.kpi_patterns = kpi_patterns

        # Set models to eval mode
        self.sentiment_model.eval()
        self.risk_model.eval()

    def predict_sentiment(self, text):
        """Predict sentiment of text (positive/negative/neutral)."""
        inputs = self.sentiment_tokenizer(
            text[:512],  # Limit to 512 tokens
            return_tensors='pt',
            truncation=True,
            padding=True
        )

        with torch.no_grad():
            outputs = self.sentiment_model(**inputs)
            logits = outputs.logits
            probabilities = torch.softmax(logits, dim=1)
            predicted_label_id = torch.argmax(logits, dim=1).item()

        # Map to label
        predicted_label = self.sentiment_labels['id_to_label'][str(predicted_label_id)]
        confidence = float(probabilities[0, predicted_label_id].item())

        return {
            'sentiment': predicted_label,
            'confidence': confidence
        }

    def predict_risk(self, text):
        """Predict risk severity of text (low/medium/high)."""
        inputs = self.risk_tokenizer(
            text[:512],
            return_tensors='pt',
            truncation=True,
            padding=True
        )

        with torch.no_grad():
            outputs = self.risk_model(**inputs)
            logits = outputs.logits
            probabilities = torch.softmax(logits, dim=1)
            predicted_label_id = torch.argmax(logits, dim=1).item()

        # Map to label
        predicted_label = self.risk_labels['id_to_label'][str(predicted_label_id)]
        confidence = float(probabilities[0, predicted_label_id].item())

        return {
            'severity': predicted_label,
            'confidence': confidence
        }

    def extract_kpis(self, text):
        """Extract KPIs from text using regex patterns."""
        extracted_kpis = {}
        text_lower = text.lower()

        for kpi_name, patterns in self.kpi_patterns.items():
            for pattern in patterns:
                matches = re.findall(pattern, text_lower, re.IGNORECASE)
                if matches:
                    extracted_kpis[kpi_name] = matches[0]
                    break

        return extracted_kpis

    def analyze_document(self, text, doc_type="unknown", company="unknown"):
        """
        Analyze complete financial document.

        Returns:
            dict: Comprehensive analysis with sentiment, risks, and KPIs
        """

        # Segment text (simple heuristic)
        segments = self._segment_text(text)

        # Analyze sentiment from CEO letter
        ceo_sentiment = None
        if segments['ceo_letter']:
            ceo_sentiment = self.predict_sentiment(segments['ceo_letter'])

        # Extract risks from risk factors section
        risks = []
        if segments['risk_factors']:
            # Split into sentences and analyze each
            sentences = re.split(r'(?<=[.!?])\s+', segments['risk_factors'][:500])
            for sentence in sentences[:5]:  # Analyze first 5 sentences
                if len(sentence) > 20:
                    risk_pred = self.predict_risk(sentence)
                    risks.append({
                        'statement': sentence.strip(),
                        'severity': risk_pred['severity'],
                        'confidence': risk_pred['confidence']
                    })

        # Extract KPIs from financial statements
        kpis = self.extract_kpis(segments['financial_statements'])

        # Compile results
        results = {
            'document_type': doc_type,
            'company': company,
            'sentiment': ceo_sentiment,
            'top_risks': sorted(risks, key=lambda x: {'high': 3, 'medium': 2, 'low': 1}.get(x['severity'], 0), reverse=True)[:5],
            'extracted_kpis': kpis,
            'analysis_complete': True
        }

        return results

    def _segment_text(self, text):
        """Simple text segmentation."""
        segments = {
            'ceo_letter': '',
            'risk_factors': '',
            'financial_statements': '',
        }

        text_lower = text.lower()

        # Find CEO letter
        for keyword in ['letter to shareholder', 'message from ceo', 'chairman\'s message']:
            idx = text_lower.find(keyword)
            if idx != -1:
                segments['ceo_letter'] = text[idx:idx + 1000]
                break

        # Find risk factors
        for keyword in ['risk factor', 'risks and uncertainties']:
            idx = text_lower.find(keyword)
            if idx != -1:
                segments['risk_factors'] = text[idx:idx + 2000]
                break

        # Find financial statements
        for keyword in ['balance sheet', 'financial statement']:
            idx = text_lower.find(keyword)
            if idx != -1:
                segments['financial_statements'] = text[idx:idx + 1000]
                break

        return segments

# ============================================================================
# PART D: Initialize Pipeline
# ============================================================================

print("ðŸ“‹ Initializing analysis pipeline...\n")

pipeline_obj = FinancialAnalysisPipeline(
    sentiment_tokenizer=sentiment_tokenizer,
    sentiment_model=sentiment_model,
    risk_tokenizer=risk_tokenizer,
    risk_model=risk_model,
    sentiment_labels=sentiment_labels,
    risk_labels=risk_labels,
    kpi_patterns=kpi_patterns
)

print("âœ“ Pipeline ready for inference")

# ============================================================================
# PART E: Test Pipeline on Sample Texts
# ============================================================================

print("\n" + "=" * 70)
print("TESTING PIPELINE ON SAMPLE TEXTS")
print("=" * 70)

test_texts = [
    {
        'text': 'Dear Shareholders, We are proud to report strong revenue growth of 20% and record profits. Our innovative products continue to drive market expansion.',
        'type': 'annual',
        'company': 'Tech Corp'
    },
    {
        'text': 'Risk Factor: The company faces significant regulatory challenges and potential litigation that could materially impact operations and profitability.',
        'type': 'quarterly',
        'company': 'Finance Ltd'
    }
]

results_list = []

for idx, test_doc in enumerate(test_texts, 1):
    print(f"\nðŸ“„ Test Document {idx}: {test_doc['company']}")
    print("-" * 70)

    result = pipeline_obj.analyze_document(
        test_doc['text'],
        doc_type=test_doc['type'],
        company=test_doc['company']
    )

    # Print results
    print(f"Company: {result['company']}")
    print(f"Document Type: {result['document_type']}")

    if result['sentiment']:
        print(f"Sentiment: {result['sentiment']['sentiment']} ({result['sentiment']['confidence']:.2%} confidence)")

    if result['top_risks']:
        print(f"Risks Found: {len(result['top_risks'])}")
        for risk in result['top_risks'][:2]:
            print(f"  - [{risk['severity'].upper()}] {risk['statement'][:60]}... ({risk['confidence']:.2%})")

    if result['extracted_kpis']:
        print(f"KPIs Found: {list(result['extracted_kpis'].keys())}")

    results_list.append(result)

# ============================================================================
# PART F: Save Pipeline & Results
# ============================================================================

print("\n" + "=" * 70)
print("âœ“ SAVING PIPELINE")
print("=" * 70)

# Save test results
results_df = pd.DataFrame([
    {
        'company': r['company'],
        'document_type': r['document_type'],
        'sentiment': r['sentiment']['sentiment'] if r['sentiment'] else 'N/A',
        'top_risk_severity': r['top_risks'][0]['severity'] if r['top_risks'] else 'N/A',
        'kpis_found': len(r['extracted_kpis'])
    }
    for r in results_list
])

results_csv = f"{inference_path}/test_inference_results.csv"
results_df.to_csv(results_csv, index=False)

print(f"âœ“ Test results saved to: {results_csv}")

# Save pipeline configuration
pipeline_config = {
    'sentiment_model': sentiment_model_path,
    'risk_model': risk_model_path,
    'kpi_patterns': str(kpi_patterns),
    'models_trained': True
}

with open(f"{inference_path}/pipeline_config.json", 'w') as f:
    json.dump(pipeline_config, f, indent=2)

print(f"âœ“ Pipeline config saved to: {inference_path}/pipeline_config.json")

# ============================================================================
# SUMMARY
# ============================================================================

print("\n" + "=" * 70)
print("âœ… INFERENCE PIPELINE CREATED & TESTED")
print("=" * 70)

print(f"""
Completed:
âœ“ Loaded both trained models (Sentiment + Risk)
âœ“ Created unified analysis pipeline
âœ“ Tested on sample financial documents
âœ“ Successfully predicted sentiment, risks, and KPIs

Pipeline Features:
- analyze_document(text) â†’ Full financial analysis
- predict_sentiment(text) â†’ Sentiment classification
- predict_risk(text) â†’ Risk severity classification
- extract_kpis(text) â†’ KPI extraction

Results Location: {inference_path}/
  - test_inference_results.csv
  - pipeline_config.json

Next: Step 7 - Build API to serve predictions
  (Create Flask/FastAPI server for dashboard integration)
""")
```

```python
# ============================================================================
# STEP 7: FLASK API SERVER (CORRECTED - LOCAL MODEL LOADING)
# ============================================================================

!pip install flask -q

import os
import json
import re
import torch
import warnings
warnings.filterwarnings('ignore')

from transformers import AutoTokenizer, AutoModelForSequenceClassification
from flask import Flask, request, jsonify
from datetime import datetime

# Paths
output_base = "/content/drive/MyDrive/processed_data"
models_path = f"{output_base}/trained_models"

print("=" * 70)
print("STEP 7: BUILDING FLASK API SERVER")
print("=" * 70)

# ============================================================================
# CHECK IF MODELS EXIST
# ============================================================================

sentiment_model_path = f"{models_path}/finbert_sentiment_final"
risk_model_path = f"{models_path}/finbert_risk_final"

print("\nðŸ” Checking model paths...")
print(f"Sentiment model path: {sentiment_model_path}")
print(f"Risk model path: {risk_model_path}")

if not os.path.exists(sentiment_model_path):
    print(f"âŒ ERROR: Sentiment model not found at {sentiment_model_path}")
    print("Make sure you ran Step 4 successfully!")
else:
    print("âœ“ Sentiment model path exists")

if not os.path.exists(risk_model_path):
    print(f"âŒ ERROR: Risk model not found at {risk_model_path}")
    print("Make sure you ran Step 5 successfully!")
else:
    print("âœ“ Risk model path exists")

# ============================================================================
# LOAD MODELS FROM LOCAL DISK
# ============================================================================

print("\nðŸ¤– Loading models from local disk...\n")

try:
    # Load sentiment model and tokenizer from local files
    print("Loading sentiment tokenizer...")
    sentiment_tokenizer = AutoTokenizer.from_pretrained(sentiment_model_path, local_files_only=True)
    print("Loading sentiment model...")
    sentiment_model = AutoModelForSequenceClassification.from_pretrained(
        sentiment_model_path,
        local_files_only=True
    )
    sentiment_model.eval()

    with open(f"{sentiment_model_path}/label_mapping.json", 'r') as f:
        sentiment_labels = json.load(f)

    print("âœ“ Sentiment model loaded successfully")
except Exception as e:
    print(f"âŒ Error loading sentiment model: {str(e)}")
    raise

try:
    # Load risk model and tokenizer from local files
    print("Loading risk tokenizer...")
    risk_tokenizer = AutoTokenizer.from_pretrained(risk_model_path, local_files_only=True)
    print("Loading risk model...")
    risk_model = AutoModelForSequenceClassification.from_pretrained(
        risk_model_path,
        local_files_only=True
    )
    risk_model.eval()

    with open(f"{risk_model_path}/label_mapping.json", 'r') as f:
        risk_labels = json.load(f)

    print("âœ“ Risk model loaded successfully")
except Exception as e:
    print(f"âŒ Error loading risk model: {str(e)}")
    raise

# Create Flask app
app = Flask(__name__)

# ============================================================================
# Prediction Functions
# ============================================================================

def predict_sentiment(text):
    """Predict sentiment of text."""
    try:
        inputs = sentiment_tokenizer(
            text[:512],
            return_tensors='pt',
            truncation=True,
            padding=True
        )

        with torch.no_grad():
            outputs = sentiment_model(**inputs)
            logits = outputs.logits
            probabilities = torch.softmax(logits, dim=1)
            predicted_label_id = torch.argmax(logits, dim=1).item()

        predicted_label = sentiment_labels['id_to_label'][str(predicted_label_id)]
        confidence = float(probabilities[0, predicted_label_id].item())

        return {'sentiment': predicted_label, 'confidence': confidence}
    except Exception as e:
        print(f"Error in predict_sentiment: {e}")
        raise

def predict_risk(text):
    """Predict risk severity of text."""
    try:
        inputs = risk_tokenizer(
            text[:512],
            return_tensors='pt',
            truncation=True,
            padding=True
        )

        with torch.no_grad():
            outputs = risk_model(**inputs)
            logits = outputs.logits
            probabilities = torch.softmax(logits, dim=1)
            predicted_label_id = torch.argmax(logits, dim=1).item()

        predicted_label = risk_labels['id_to_label'][str(predicted_label_id)]
        confidence = float(probabilities[0, predicted_label_id].item())

        return {'severity': predicted_label, 'confidence': confidence}
    except Exception as e:
        print(f"Error in predict_risk: {e}")
        raise

def extract_kpis(text):
    """Extract KPIs from text using regex."""
    kpi_patterns = {
        'revenue': [r'revenue.*?([\$â‚¹]\s*[\d,\.]+\s*(?:billion|million|crore|lakh))'],
        'net_profit': [r'net profit.*?([\$â‚¹]\s*[\d,\.]+\s*(?:billion|million|crore|lakh))'],
        'eps': [r'earnings per share.*?([\d\.]+)'],
        'roe': [r'return on equity.*?([\d\.]+)\s*%'],
    }

    extracted = {}
    text_lower = text.lower()

    for kpi_name, patterns in kpi_patterns.items():
        for pattern in patterns:
            matches = re.findall(pattern, text_lower, re.IGNORECASE)
            if matches:
                extracted[kpi_name] = matches[0]
                break

    return extracted

# ============================================================================
# API Routes
# ============================================================================

@app.route('/health', methods=['GET'])
def health_check():
    """Health check endpoint."""
    return jsonify({'status': 'OK', 'message': 'API is running'}), 200

@app.route('/api/v1/sentiment', methods=['POST'])
def analyze_sentiment():
    """Analyze sentiment of text."""
    try:
        data = request.get_json()
        if not data or 'text' not in data:
            return jsonify({'error': 'Missing text in request'}), 400

        text = data['text']
        company = data.get('company', 'Unknown')

        result = predict_sentiment(text)

        return jsonify({
            'sentiment': result['sentiment'],
            'confidence': float(result['confidence']),
            'company': company,
            'timestamp': datetime.now().isoformat(),
            'status': 'success'
        }), 200
    except Exception as e:
        return jsonify({'error': str(e)}), 500

@app.route('/api/v1/risk', methods=['POST'])
def analyze_risk():
    """Analyze risk severity of text."""
    try:
        data = request.get_json()
        if not data or 'text' not in data:
            return jsonify({'error': 'Missing text in request'}), 400

        text = data['text']
        company = data.get('company', 'Unknown')

        result = predict_risk(text)

        return jsonify({
            'severity': result['severity'],
            'confidence': float(result['confidence']),
            'company': company,
            'timestamp': datetime.now().isoformat(),
            'status': 'success'
        }), 200
    except Exception as e:
        return jsonify({'error': str(e)}), 500

@app.route('/api/v1/kpis', methods=['POST'])
def extract_document_kpis():
    """Extract KPIs from financial text."""
    try:
        data = request.get_json()
        if not data or 'text' not in data:
            return jsonify({'error': 'Missing text in request'}), 400

        text = data['text']
        company = data.get('company', 'Unknown')

        kpis = extract_kpis(text)

        return jsonify({
            'kpis': kpis,
            'kpis_found': len(kpis),
            'company': company,
            'timestamp': datetime.now().isoformat(),
            'status': 'success'
        }), 200
    except Exception as e:
        return jsonify({'error': str(e)}), 500

@app.route('/api/v1/analyze', methods=['POST'])
def full_analysis():
    """Perform full financial document analysis."""
    try:
        data = request.get_json()
        if not data or 'text' not in data:
            return jsonify({'error': 'Missing text in request'}), 400

        text = data['text']
        company = data.get('company', 'Unknown')
        doc_type = data.get('document_type', 'unknown')

        # Analyze sentiment
        sentiment = predict_sentiment(text)

        # Extract KPIs
        kpis = extract_kpis(text)

        # Extract risks (analyze sentences)
        sentences = re.split(r'(?<=[.!?])\s+', text[:1000])
        risks = []
        for sentence in sentences[:5]:
            if len(sentence) > 20:
                risk_pred = predict_risk(sentence)
                risks.append({
                    'statement': sentence[:100],
                    'severity': risk_pred['severity'],
                    'confidence': float(risk_pred['confidence'])
                })

        # Sort risks by severity
        severity_order = {'high': 3, 'medium': 2, 'low': 1}
        risks = sorted(risks, key=lambda x: severity_order.get(x['severity'], 0), reverse=True)

        return jsonify({
            'company': company,
            'document_type': doc_type,
            'sentiment': sentiment,
            'risks': risks[:3],
            'kpis': kpis,
            'timestamp': datetime.now().isoformat(),
            'status': 'success'
        }), 200
    except Exception as e:
        return jsonify({'error': str(e)}), 500

# ============================================================================
# Start Server
# ============================================================================

if __name__ == '__main__':
    print("\n" + "=" * 70)
    print("âœ… FLASK API SERVER READY")
    print("=" * 70)

    print(f"""
API Endpoints:
1. GET /health
2. POST /api/v1/sentiment
3. POST /api/v1/risk
4. POST /api/v1/kpis
5. POST /api/v1/analyze

Server: http://localhost:5000
Starting server...
""")

    app.run(host='0.0.0.0', port=5000, debug=False)
```
